### Comparing lemma frequency to hofstede scores

```python
# shared_verbs_vertical = utils.make_shared(verb_counts,nations_ofinterest20_ordered,vertical=True).to_csv(save_progress+"shared_verbs_vertical.csv")
shared_verbs_vertical = pd.read_csv(save_progress+"shared_verbs_vertical.csv")

# shared_nouns_vertical = utils.make_shared(noun_counts,nations_ofinterest20_ordered,vertical=True).to_csv(save_progress+"shared_nouns_vertical.csv")
shared_nouns_vertical = pd.read_csv(save_progress+"shared_nouns_vertical.csv")

verb_hofstede = shared_verbs_vertical.merge(hofstede[['pdi','idv','mas','uai','ltowvs','ivr','main']], left_on='nation', right_on='main', how='left').drop(columns='main')
noun_hofstede = shared_nouns_vertical.merge(hofstede[['pdi','idv','mas','uai','ltowvs','ivr','main']], left_on='nation', right_on='main', how='left').drop(columns='main')
lemma_hofstede = pd.concat([verb_hofstede,noun_hofstede])
```

```python
lemma_min = lemma_hofstede[['lemma','portion']].copy().groupby('lemma').min().reset_index()
lemma_max = lemma_hofstede[['lemma','portion']].copy().groupby('lemma').max().reset_index()
lemma_diff = lemma_max.merge(lemma_min, on='lemma', how='left', suffixes=['_max','_min'])
lemma_diff['diff'] = lemma_diff['portion_max'] - lemma_diff['portion_min']
lemma_diff = dict(zip(lemma_diff['lemma'],lemma_diff['diff']))

lemma_hofstede['diff'] = lemma_hofstede['lemma'].map(lemma_diff)
```

```python
lemma_diff90 = np.percentile(lemma_hofstede['diff'],90)
relativewords = {'hero','mother','father','king','emperor','dragon'}
lemma_hofstede_plot = lemma_hofstede.copy().query('lemma in @relativewords')
lemma_hofstede_plot
```

```python
summarizer = pipeline("summarization")
```

```python
top_stories_by_topic_summarized = pd.DataFrame(columns=['topic','nation','title','summary'])
for topic in top_stories_by_topic.keys():
    for i in top_stories_by_topic[topic]:
        story = stories.iloc[i]['text']
        try:
            summarized = summarizer(story, max_length=100, min_length=30)
            summarized = summarizer(story, max_length=100, min_length=30)[0]['summary_text']
            top_stories_by_topic_summarized.loc[len(top_stories_by_topic_summarized)] = [topic, stories.iloc[i]['nation'], stories.iloc[i]['title'], summarized]
        except IndexError:
            pass
top_stories_by_topic_summarized.to_csv(save_progress+'top_stories_by_topic_summarized.csv')
top_stories_by_topic_summarized
```
